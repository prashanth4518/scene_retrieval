```python
def closure(i,results):
    done_indices = []
    lis_old = []
    lis_new = []
    l = results[i]
    lis_new = lis_old + list(l)
    done_indices.append(i)
    
    while(list(set(lis_new)-set(lis_old)) != []):
        a = list(set(lis_new) - set(done_indices))
        if a == []:
            return set([i])

        else:
            a = a[0]
        l = results[a]
        
        lis_old = lis_new
        lis_new = lis_old + list(l)
        done_indices.append(a)
        
    return set(lis_new)
```
